```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import healpy as hp
```

# 1M

```python
NSIDE = 512
print(
    "Approximate resolution at NSIDE {} is {:.2} deg".format(
        NSIDE, hp.nside2resol(NSIDE, arcmin=True) / 60
    )
)
```

```python
NPIX = hp.nside2npix(NSIDE)
print(NPIX)
```

```python
ori = np.loadtxt("./frequency_1M/FF_nest_1m.txt")
```

```python
hp.mollview(ori, nest=True, title="Mollview image B with NSIDE=512")
hp.graticule()  # 
plt.show()
```

```python
recon_1m = np.loadtxt("./F_recon_1M/F1_healpix.txt")
```

```python
hp.mollview(recon_1m, nest=True, title="Mollview Recon Image with NSIDE=512")
hp.graticule()  # 
plt.show()
```

```python
total_data = np.loadtxt("./F_recon_1M/F1_healpix.txt")

# 450
for i in range(2, 51):
    file_path = f"./F_recon_1M/F{i}_healpix.txt"
    data = np.loadtxt(file_path)
    
    # nan0
    data = np.nan_to_num(data)
    
    # 
    total_data += data
    
hp.mollview(total_data, nest=True, title="Mollview Recon image healpix Blockage with NSIDE=512 of 50 Periods")
hp.graticule()  # 
plt.show()
```
